```python
import random
numbers = []
counter = 1
total = 0
while counter <= 50:
    num = random.randint(25,40)
    numbers.append(num)
    counter = counter + 1
total = sum(numbers)
print(f'The total sum of the generated 50 numbers is {total}')
```
