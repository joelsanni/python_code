# generat a random number between 45,70 and this numbers should not be less then 35
# modify this numbers by finding the sqrt of each 

```python
import random 
number = []
counter = 1
while counter <=35:
    num = random.randint(45,70)
    number.append(num)
    counter += 1
```

```python
import math
u_numbers = list(map(lambda digits:math.sqrt(digits),number))
print(u_numbers)
```
